```python
stopwords = None #set(STOPWORDS)
#stopwords.add("++")
def show_wordcloud(data, title = None):
    wordcloud = WordCloud(
        background_color='white',
        stopwords=stopwords,
        regexp=r'[a-zA-z0-9\+\-]+',
        max_words=50,
        max_font_size=40, 
        scale=5,
        random_state=1
    ).generate(" ".join(data))

    fig = plt.figure(1, figsize=(25, 25))
    plt.axis('off')
    
    if title: 
        fig.suptitle(title, fontsize=20)
        fig.subplots_adjust(top=2.3)

    plt.imshow(wordcloud)
    plt.show()
```

```python
from IPython.display import HTML, display

words = [
    ('Java', 52, 'lang'), ('Spring-Boot', 44, 'lang'), ('Hibernate', 36, 'lang'),
    ('Python', 52, 'lang'), ('Django', 44, 'framework'), ('Flask', 44, 'framework'),
    ('Virtualization', 36, 'infra'), ('Docker', 52, 'infra'), ('Kubernetes', 52, 'infra'),
    ('PHP', 44, 'lang'), ('Codeigniter', 36, 'framework'),
    ('AWS', 44, 'infra'), ('Git', 44, 'infra'),
    ('SQL', 44, 'data'), ('MySQL', 44, 'data'), ('PostgreSQL', 44, 'data'),
    ('MongoDB', 44, 'data'), ('JSON', 36, 'data'),
    ('REST', 36, 'web'), ('API', 36, 'web'), ('HTML', 36, 'web'),
    ('CSS', 36, 'web'), ('JavaScript', 44, 'web'),
    ('C', 44, 'lang'), ('C++', 44, 'lang'),
    ('IoT', 36, 'other'), ('AI', 44, 'ml'), ('ML', 44, 'ml'), ('DL', 36, 'ml'),
    ('Linux', 44, 'infra'), ('bash', 44, 'infra'), ('shell', 36, 'infra'),
]

CATEGORY_COLORS = {
    'lang':      ('#dbeafe', '#1e40af'),
    'framework': ('#dcfce7', '#166534'),
    'infra':     ('#ffe4d6', '#9a3412'),
    'data':      ('#ede9fe', '#4c1d95'),
    'ml':        ('#fef3c7', '#92400e'),
    'web':       ('#f1f5f9', '#334155'),
    'other':     ('#f1f5f9', '#334155'),
}

SIZE_TO_FONT = {52: 28, 44: 22, 36: 16}

items_html = ""
for word, size, cat in words:
    bg, fg = CATEGORY_COLORS[cat]
    fs = SIZE_TO_FONT[size]
    items_html += f"""
    <div style="display:inline-flex; align-items:center; justify-content:center;
                background:{bg}; color:{fg};
                border-radius:8px; padding:6px 14px; margin:5px;
                font-size:{fs}px; font-weight:500; font-family:sans-serif;
                cursor:default;">
      {word}
    </div>"""

legend_html = ""
labels = [('Languages','#dbeafe','#1e40af'), ('Frameworks','#dcfce7','#166534'),
          ('Infra/DevOps','#ffe4d6','#9a3412'), ('Data/DB','#ede9fe','#4c1d95'),
          ('AI/ML','#fef3c7','#92400e'), ('Web/Other','#f1f5f9','#334155')]
for label, bg, fg in labels:
    legend_html += f'<span style="background:{bg};color:{fg};border-radius:20px;padding:3px 10px;margin:3px;font-size:12px;font-family:sans-serif;">{label}</span>'

html = f"""
<div style="border:1px solid #e2e8f0; border-radius:12px; padding:20px; background:white;">
  <p style="text-align:center; font-size:20px; font-weight:600;
            font-family:sans-serif; color:#1e293b; margin-bottom:4px;">
    ... Technologies ...
  </p>
  <div style="text-align:center; margin-bottom:14px;">{legend_html}</div>
  <div style="display:flex; flex-wrap:wrap; justify-content:center; align-items:center;">
    {items_html}
  </div>
</div>
"""

display(HTML(html))
```
